```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

##### Generate data

```python
np.random.seed(54321)

k = 10
X = np.random.normal(size = (1000, k))
X = np.hstack((X,np.ones((X.shape[0],1))))

# True beta:
beta = np.random.normal(size = k + 1) # +1 for intercept term
lam = np.exp(np.dot(X, beta))

y = np.random.poisson(lam = lam)
```

##### Estimate Poisson GLM

```python
# No closed form; so we use gradient descent 

epochs = 3**5
eta = 0.00001
beta_hat = np.random.normal(size = k + 1)


def log_lik(beta_hat, X, y):
    return np.sum(np.exp(np.dot(X, beta_hat)) - y*np.dot(X, beta_hat))

losses = []
for epoch in range(epochs):
    yhat = np.exp(np.dot(X, beta_hat))

    # Compute loss
    losses.append(log_lik(beta_hat, X, y))

    # Compute gradient:
    grad = np.dot(X.T, yhat - y)
    #print(grad)
    beta_hat -= eta*grad
    
```

##### Plot fit:

```python
fig, ax = plt.subplots()
sns.scatterplot(beta, beta_hat)
ax.set(xlabel = 'True Betas', ylabel = 'Estimated Betas', title = 'Beta vs. Beta Hats');
```

```python
fig, ax = plt.subplots()
sns.lineplot(np.arange(len(losses)), losses)
ax.set(xlabel = 'Iteration', ylabel = 'Log loss', title = 'Log loss');
```
