```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import seaborn as sns
```

##### Data generation

```python
# Generate data as 3 clusters
data = make_blobs(n_samples = 1000, n_features = 2, centers = 3, random_state = 50)

# Plot generated clusters
ax = sns.scatterplot(data[0][:,0], data[0][:,1], hue = data[1])
```

##### Implementation of K-means

```python
num_clusters = 3

def init_clusters(k, data):
    # Select k points at random as clusters
    clusters = data[np.random.choice(data.shape[0], k, replace = False)]
    return clusters 

def find_nearest_centroids(data, centroids):
    # Find distance between each point and centroids
    distances = []
    for centroid in centroids:
        distances.append(np.sum(np.power(data - centroid, 2), axis = 1))
    distances = np.vstack(distances).T
    
    # Get closest centroid to each point
    closest = np.argmin(distances, axis = 1)
    return closest
    
def compute_centroids(data, assignments, k):
    centroids = np.zeros((k, data.shape[1]))
    # For each cluster, compute average of datapoints in that cluster
    for j in range(k):
        centroids[j, :] = (np.mean(data[assignments == j], axis = 0))
    return centroids

def k_means(data, k = 3, max_iters = 1000):
    # Randomly initialize centroids
    old_centroids = init_clusters(k, data)
    
    # While clusters are un-changing, iteratively find better clusters:
    for iter in range(max_iters):
        assignments = find_nearest_centroids(data, old_centroids)
        new_centroids = compute_centroids(data, assignments, k)
      
        if np.all(old_centroids == new_centroids):
            break
        else:
            old_centroids = new_centroids
    
    return assignments
        
# Run the algorithm    
assignments = k_means(data = data[0])

# Plot found clusters
ax = sns.scatterplot(data[0][:,0], data[0][:,1], hue = assignments)
    
```
